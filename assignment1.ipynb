```python
#importing modules
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#plot a line between a and b:
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b, 'ro-', linewidth =2, markersize=5)
plt.title('question 1 Solution')
plt.xlabel('x-axis : a')
plt.ylabel('y-axis : b')
plt.show()
```

```python
#plot a line showing the sales trend in company 1 and 2:
days=[1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180]
sales_2 =[70,90,160,150,140,145,175]
plt.plot(days,sales_1,'g--',label="sales1")
plt.plot(days,sales_2,'r--',label="sales2")
plt.xlabel('x-axis : days')
plt.ylabel('y-axis : sales')
plt.legend()
plt.title('Question 2 Solutin : The sales trend in company 1 and 2')
plt.show()
```

```python
#create a 3 by 3 subplots
x=[1,2,3,4]
y1=[4,3,2,1]
y2=[10,20,30,40]
y3=[40,30,20,10]
y4=[1,2,1,2]
y5=[40,70,90,70]

plt.figure(figsize = (15,15))

plt.subplot(3,3,1)
plt.plot(x,y1,'r--')
plt.xlabel('x')
plt.ylabel('y1')
plt.title('(x,y1)')

plt.subplot(3,3,2)
plt.plot(x,y2,'b--')
plt.xlabel('x')
plt.ylabel('y2')
plt.title('(x,y2)')

plt.subplot(3,3,3)
plt.plot(x,y3,'g--')
plt.xlabel('x')
plt.ylabel('y3')
plt.title('(x,y3)')

plt.subplot(3,3,4)
plt.plot(x,y4,'bo-')
plt.xlabel('x')
plt.ylabel('y4')
plt.title('(x,y4)')

plt.subplot(3,3,5)
plt.plot(x,y5,'ro-')
plt.xlabel('x')
plt.ylabel('y5')
plt.title('(x,y5)')
```
